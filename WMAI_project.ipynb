```python
# Step 9: Stock Selection Analysis
print("STOCK SELECTION ANALYSIS")
print("="*40)

# Create a comprehensive scoring system
def calculate_composite_scores(df):
    """
    Calculate composite scores for each bank based on multiple financial metrics
    """
    # Define weights for different categories
    category_weights = {
        'Profitability': 0.35,
        'Efficiency': 0.25,
        'Solvency': 0.20,
        'Valuation': 0.20
    }

    # Define metrics and their weights within each category
    metric_weights = {
        # Profitability metrics
        'ROE (%)': 0.30,
        'ROA (%)': 0.25,
        'Net Profit Margin (%)': 0.25,
        'Net Interest Margin (X)': 0.20,

        # Efficiency metrics
        'Cost to Income (%)': 0.40,
        'CASA (%)': 0.35,
        'Asset Turnover Ratio': 0.25,

        # Solvency metrics
        'Capital Adequacy Ratio (%)': 0.60,
        'Debt to Equity': 0.40,

        # Valuation metrics
        'P/E Ratio': 0.40,
        'P/B Ratio': 0.35,
        'Dividend/Share (Rs.)': 0.25
    }

    # Normalize and score each metric
    scores_df = df[df['Year'] == '2025'].copy()

    # Higher is better metrics (profitability, efficiency, solvency)
    higher_better = ['ROE (%)', 'ROA (%)', 'Net Profit Margin (%)', 'Net Interest Margin (X)',
                    'CASA (%)', 'Asset Turnover Ratio', 'Capital Adequacy Ratio (%)', 'Dividend/Share (Rs.)']

    # Lower is better metrics
    lower_better = ['Cost to Income (%)', 'Debt to Equity', 'P/E Ratio', 'P/B Ratio']

    # Calculate normalized scores for each metric
    for metric in higher_better + lower_better:
        if metric in scores_df.columns:
            if metric in higher_better:
                # Normalize with higher values being better
                max_val = scores_df[metric].max()
                min_val = scores_df[metric].min()
                if max_val != min_val:
                    scores_df[f'{metric}_Score'] = (scores_df[metric] - min_val) / (max_val - min_val) * 100
                else:
                    scores_df[f'{metric}_Score'] = 100  # All equal
            else:
                # Normalize with lower values being better
                max_val = scores_df[metric].max()
                min_val = scores_df[metric].min()
                if max_val != min_val:
                    scores_df[f'{metric}_Score'] = (max_val - scores_df[metric]) / (max_val - min_val) * 100
                else:
                    scores_df[f'{metric}_Score'] = 100  # All equal

    # Calculate category scores
    scores_df['Profitability_Score'] = (
        scores_df['ROE (%)_Score'] * metric_weights['ROE (%)'] +
        scores_df['ROA (%)_Score'] * metric_weights['ROA (%)'] +
        scores_df['Net Profit Margin (%)_Score'] * metric_weights['Net Profit Margin (%)'] +
        scores_df['Net Interest Margin (X)_Score'] * metric_weights['Net Interest Margin (X)']
    )

    scores_df['Efficiency_Score'] = (
        scores_df['Cost to Income (%)_Score'] * metric_weights['Cost to Income (%)'] +
        scores_df['CASA (%)_Score'] * metric_weights['CASA (%)'] +
        scores_df['Asset Turnover Ratio_Score'] * metric_weights['Asset Turnover Ratio']
    )

    scores_df['Solvency_Score'] = (
        scores_df['Capital Adequacy Ratio (%)_Score'] * metric_weights['Capital Adequacy Ratio (%)'] +
        scores_df['Debt to Equity_Score'] * metric_weights['Debt to Equity']
    )

    scores_df['Valuation_Score'] = (
        scores_df['P/E Ratio_Score'] * metric_weights['P/E Ratio'] +
        scores_df['P/B Ratio_Score'] * metric_weights['P/B Ratio'] +
        scores_df['Dividend/Share (Rs.)_Score'] * metric_weights['Dividend/Share (Rs.)']
    )

    # Calculate overall composite score
    scores_df['Composite_Score'] = (
        scores_df['Profitability_Score'] * category_weights['Profitability'] +
        scores_df['Efficiency_Score'] * category_weights['Efficiency'] +
        scores_df['Solvency_Score'] * category_weights['Solvency'] +
        scores_df['Valuation_Score'] * category_weights['Valuation']
    )

    return scores_df

# Calculate composite scores
scored_banks = calculate_composite_scores(df_all_banks)

# Rank banks by composite score
scored_banks['Overall_Rank'] = scored_banks['Composite_Score'].rank(ascending=False, method='min')
scored_banks = scored_banks.sort_values('Overall_Rank')

# Display the scoring results
print("\nCOMPREHENSIVE SCORING RESULTS - ALL BANKS (2025):")
display_cols = ['Bank', 'Overall_Rank', 'Composite_Score', 'Profitability_Score',
                'Efficiency_Score', 'Solvency_Score', 'Valuation_Score']
display(scored_banks[display_cols].set_index('Bank').style.background_gradient(cmap='RdYlGn', axis=0))

# Visualize the category scores
category_scores = scored_banks[['Bank', 'Profitability_Score', 'Efficiency_Score',
                               'Solvency_Score', 'Valuation_Score']].set_index('Bank')

plt.figure(figsize=(14, 8))
category_scores.plot(kind='bar', stacked=False, colormap='Set2')
plt.title('Category-wise Performance Scores by Bank (2025)')
plt.ylabel('Score (0-100)')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Step 10: Final Stock Selection and Justification
print("FINAL STOCK SELECTION AND JUSTIFICATION")
print("="*50)

# Select top 3 banks
top_3_banks = scored_banks.head(3)

print("\nTOP 3 RECOMMENDED STOCKS (STRONG BUY):")
print("-" * 50)

for idx, (_, bank) in enumerate(top_3_banks.iterrows(), 1):
    print(f"\n{idx}. {bank['Bank']} ({bank['Ticker']})")
    print(f"   Overall Rank: {int(bank['Overall_Rank'])}")
    print(f"   Composite Score: {bank['Composite_Score']:.2f}")
    print(f"   Key Strengths:")

    # Identify top 3 strengths
    strengths = []
    if bank['Profitability_Score'] >= 80:
        strengths.append("High Profitability")
    if bank['Efficiency_Score'] >= 80:
        strengths.append("Operational Efficiency")
    if bank['Solvency_Score'] >= 80:
        strengths.append("Strong Solvency")
    if bank['Valuation_Score'] >= 80:
        strengths.append("Attractive Valuation")

    for i, strength in enumerate(strengths[:3], 1):
        print(f"     {i}. {strength}")

    # Print key financial metrics
    print(f"   Key Financial Metrics (2025):")
    print(f"     - ROE: {bank['ROE (%)']:.2f}%")
    print(f"     - ROA: {bank['ROA (%)']:.2f}%")
    print(f"     - Net Profit Margin: {bank['Net Profit Margin (%)']:.2f}%")
    print(f"     - P/B Ratio: {bank['P/B Ratio']:.2f}")
    print(f"     - CASA Ratio: {bank['CASA (%)']:.2f}%")

# Compare public vs private sector banks
private_banks = ['HDFC Bank', 'ICICI Bank', 'Axis Bank']
public_banks = ['SBI', 'PNB', 'BOI']

private_avg_score = scored_banks[scored_banks['Bank'].isin(private_banks)]['Composite_Score'].mean()
public_avg_score = scored_banks[scored_banks['Bank'].isin(public_banks)]['Composite_Score'].mean()

print(f"\nSECTOR ANALYSIS:")
print(f"Private Banks Average Score: {private_avg_score:.2f}")
print(f"Public Sector Banks Average Score: {public_avg_score:.2f}")
print(f"Performance Gap: {private_avg_score - public_avg_score:.2f} points")

# Investment strategy recommendations
print(f"\nINVESTMENT STRATEGY RECOMMENDATIONS:")
print("1. Focus on private sector banks for superior profitability and efficiency")
print("2. Consider public sector banks for dividend income and valuation opportunities")
print("3. Maintain a balanced portfolio with 60-70% allocation to top private banks")
print("4. Use any market corrections to accumulate positions in high-quality banks")

# Risk factors
print(f"\nKEY RISK FACTORS TO MONITOR:")
print("1. Interest rate changes by RBI")
print("2. Asset quality and NPA trends")
print("3. Economic growth and credit demand")
print("4. Regulatory changes affecting banking operations")
print("5. Competition from fintech and digital banking platforms")

# Create a summary table of the top 3 recommendations
recommendation_summary = top_3_banks[['Bank', 'Ticker', 'ROE (%)', 'ROA (%)',
                                     'Net Profit Margin (%)', 'P/B Ratio',
                                     'CASA (%)', 'Composite_Score']].copy()
recommendation_summary['Recommendation'] = 'Strong Buy'
recommendation_summary['Target Allocation'] = ['40%', '35%', '25%']

print("\nPORTFOLIO ALLOCATION RECOMMENDATION:")
display(recommendation_summary.set_index('Bank').style.background_gradient(cmap='RdYlGn', axis=0))
```
